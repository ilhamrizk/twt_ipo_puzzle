# BHAR250

```python
ipolist = df_fileipo[['kode', 'tanggal_ipo', 'tanggal_ipo_offset250']].T.to_dict()
is_first = True
for ipo in ipolist.values():
  kode = ipo['kode']+'.JK'
  startdate = ipo['tanggal_ipo']
  enddate = ipo['tanggal_ipo_offset250']
  if is_first:
    df = yf.Ticker(kode).history(start=startdate, end=enddate).reset_index()
    df['kode'] = ipo['kode']
    df['Volume'] = df['Volume'].fillna(0).astype(np.int64)
    is_first = False
  else:
    df_temp = yf.Ticker(kode).history(start=startdate, end=enddate).reset_index()
    df_temp['kode'] = ipo['kode']
    df_temp['Volume'] = df_temp['Volume'].fillna(0).astype(np.int64)
    df = pd.concat([df, df_temp])
```

```python
DESTFILE = '/content/drive/MyDrive/Freelance/twitter_sentiment_analysis/data_new/'
df.to_csv(DESTFILE+'data_stock_prices.csv')
```

# Market Condition

```python
filter_sektor = ['Pulp & Paper', 
                 'Technology',
                 'Consumer Goods Industry',
                 'Transportation']
df_market_cond = df_fileipo[['Kode',
                             'Sektor',
                             'tanggal_ipo']]
df_market_cond['Sektor'] = df_market_cond['Sektor'].str.strip()
df_market_cond = df_market_cond[~df_market_cond['Sektor'].isin(filter_sektor)]
tgl_ipo_dttm = pd.to_datetime(df_market_cond['tanggal_ipo'], format="%Y-%m-%d")
df_market_cond['ipo_h_1'] = tgl_ipo_dttm - pd.to_timedelta(1, unit='d')
df_market_cond['ipo_h_7'] = tgl_ipo_dttm - pd.to_timedelta(7, unit='d')
df_market_cond['ipo_h_59'] = tgl_ipo_dttm - pd.to_timedelta(59, unit='d')
df_market_cond['ipo_h_66'] = tgl_ipo_dttm - pd.to_timedelta(66, unit='d')
df_market_cond.head()
```

```python
df_reff = pd.read_csv(FOLDER+'references_market_condition.csv')\
  .drop(columns='Unnamed: 0')
df_reff.head()
```

```python
df_ = df_market_cond.merge(df_reff, 
                           left_on = 'Sektor',
                           right_on = 'Sector',
                           how = 'left')
df_.head()
```

```python
select_col = ['Kode',
              'tanggal_ipo',
              'ipo_h_7',
              'ipo_h_59',
              'ipo_h_66',
              'Kode Perusahaan']
list_kode = df_[select_col].T.to_dict()

is_first = True
for k, v in list_kode.items():
  kode = v['Kode Perusahaan']+'.JK'
  startdate_close = str(v['ipo_h_7'])[0:10]
  enddate_close = v['tanggal_ipo']

  startdate_open = str(v['ipo_h_66'])[0:10]
  enddate_open = str(v['ipo_h_59'])[0:10]
  if is_first:
    df_close = yf.Ticker(kode).history(start=startdate_close, end=enddate_close).reset_index()
    df_close['kode_reff'] = v['Kode Perusahaan']
    df_close['kode'] = v['Kode']
    df_close['Volume'] = df_close['Volume'].fillna(0).astype(np.int64)

    df_open = yf.Ticker(kode).history(start=startdate_open, end=enddate_open).reset_index()
    df_open['kode_reff'] = v['Kode Perusahaan']
    df_open['kode'] = v['Kode']
    df_open['Volume'] = df_open['Volume'].fillna(0).astype(np.int64)
    is_first = False
  else:
    df_temp_c = yf.Ticker(kode).history(start=startdate_close, end=enddate_close).reset_index()
    df_temp_c['kode_reff'] = v['Kode Perusahaan']
    df_temp_c['kode'] = v['Kode']
    df_temp_c['Volume'] = df_temp_c['Volume'].fillna(0).astype(np.int64)
    df_close = pd.concat([df_close, df_temp_c])

    df_temp_o = yf.Ticker(kode).history(start=startdate_open, end=enddate_open).reset_index()
    df_temp_o['kode_reff'] = v['Kode Perusahaan']
    df_temp_o['kode'] = v['Kode']
    df_temp_o['Volume'] = df_temp_o['Volume'].fillna(0).astype(np.int64)
    df_open = pd.concat([df_open, df_temp_o])

df_open.to_csv(FOLDER+'data_market_coditio_oe.csv')
df_close.to_csv(FOLDER+'data_market_coditio_close.csv')
```
